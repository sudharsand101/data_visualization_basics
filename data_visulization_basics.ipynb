```python
# importing matplot library
import matplotlib.pyplot as plt
# importing seaborn library
import seaborn as sns
# importing nump
import numpy as np
```

```python
# 1)scatter plot
print("1.scatter plot:")
year = np.array([2008,2010,2012,2014,2016,2018,2020,2022]) #)x column
data = np.array([500,1000,1500,2000,2500,3000,3200,4000])  #)y column

plt.xlabel("year") #) x column label
plt.ylabel("data") #) y column label
plt.title("Data vs Year") #)title

plt.scatter(year,data)
plt.show()
```

```python
#) loading the data set'flights' from kaggle
df = sns.load_dataset('flights')

#2)multiple scatter plot
print("2.multiple scatter plot")

plt.figure(figsize=(7,5))
sns.scatterplot(data=df,x="passengers",y="year",hue='month')

plt.show()
```

```python
# 3)line plot
print("3.line plot:")

year = np.array([2008,2010,2012,2014,2016,2018,2020,2022]) #)x column
data = np.array([500,1000,1500,2000,2500,3000,3200,4000])  #)y column

plt.xlabel("year")
plt.ylabel("data")
plt.title("Year vs Data")

plt.plot(year,data)
plt.show()
```

```python
# 4)multiple line plot
print("4.multiple line plot:")

year = np.array([2008,2010,2012,2014,2016,2018,2020,2022]) #)x column
data_type1 = np.array([500,1000,1500,2000,2500,3000,3200,4000])  #)y column
data_type2 = np.array([300,600,900,1200,1400,1700,2000,2600])

plt.xlabel("year")
plt.ylabel("data")
plt.title("Year vs Data")

plt.plot(year,data_type1,year,data_type2)
plt.show()
```

```python
# 5)vertical bar plot
print("5.vertical bar plot:")

year = [2008,2010,2012,2014,2016,2018,2020,2022] #)x data can be given as list
data = [500,1000,1500,2000,2500,3000,3200,4000]  #)y column

plt.xlabel("year")
plt.ylabel("data")
plt.title("Year vs Data")

plt.bar(year,data)
plt.show()
```

```python
# 6)vertical bar plot for 2 different type data
print("6.vertical bar plot:")

month = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG'] #)x data can be given as list
data_type1 = [500,1000,1500,2000,2500,3000,3200,4000]  #)y column
data_type2 = [300,600,900,1200,1400,1700,2000,2600]  #)y column

plt.xlabel("month")
plt.ylabel("data")
plt.title("month vs data")

plt.bar(month,data_type1)
plt.bar(month,data_type2,bottom = data_type2)
plt.legend(["Data type 1", "Data type 2"])

plt.show()
```

```python
# 7)horizontal bar plot
print("7.horizontal bar plot:")

year = np.array([2008,2010,2012,2014,2016,2018,2020,2022]) #)x column
data = np.array([500,1000,1500,2000,2500,3000,3200,4000])  #)y column

plt.xlabel("data")
plt.ylabel("year")
plt.title("Data vs Year")

plt.barh(year,data)
plt.show()
```

```python
# 8)piechart
print("8.pie chart:")

month = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG']
data_type1 = [500,1000,1500,2000,2500,3000,3200,4000] 

plt.pie(data_type1,labels=month)
plt.show()
```

```python
#) 9.box plot
print("9.box plot:")
sns.boxplot(df,x='month',y='passengers')
plt.show()
```
